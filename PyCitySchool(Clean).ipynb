```python
# Add names to our bins
spending_bins = [0, 585, 630, 645, 675]
group_names = ["<$584","$585-$629", "$630-$644","645-$675"]

# Categorize spending based on the bins
per_school_summary_df["Spending Range (Per Student)"] = pd.cut(per_student_budget, spending_bins, labels=group_names)
per_school_summary_df
```

```python
spending_math_scores = per_school_summary_df.groupby(["Spending Range (Per Student)"]).mean()["Average Math"]
spending_reading_scores = per_school_summary_df.groupby(["Spending Range (Per Student)"]).mean()["Average Reading"]
spending_math_percentage = per_school_summary_df.groupby(["Spending Range (Per Student)"]).mean()["% Passing Math"]  
spending_reading_percentage = per_school_summary_df.groupby(["Spending Range (Per Student)"]).mean()["% Passing Reading"]
spending_overall_percentage = per_school_summary_df.groupby(["Spending Range (Per Student)"]).mean()["% Overall Passing"]                                                                                             
```

```python
spending_overall_percentage
```

```python
# Create a new data frame
spending_summary_df = pd.DataFrame(
{"Average Math":spending_math_scores, "Average Reading":spending_reading_scores, 
 "Math Percentage":spending_math_percentage, "Reading Percentage":spending_reading_percentage,
"Overall Percentage":spending_overall_percentage })

spending_summary_df["Average Math"] = spending_summary_df["Average Math"].map("{:,.1f}".format)
spending_summary_df["Average Reading"] = spending_summary_df["Average Reading"].map("{:,.1f}".format)
spending_summary_df["Math Percentage"] = spending_summary_df["Math Percentage"].map("{:,.0f}".format)
spending_summary_df["Reading Percentage"] = spending_summary_df["Reading Percentage"].map("{:,.0f}".format)
spending_summary_df["Overall Percentage"] = spending_summary_df["Overall Percentage"].map("{:,.0f}".format)

spending_summary_df
```

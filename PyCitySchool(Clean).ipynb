```python
import pandas as pd
#Files to Load
import csv
import os
school_data_to_load = os.path.join("Resources", "schools_complete.csv")
student_data_to_load = os.path.join("Resources", "clean_students_complete.csv")
```

```python
school_data_df = pd.read_csv(school_data_to_load)
student_data_df = pd.read_csv(student_data_to_load)
```

```python
# Combine the data into a single dataset.
school_data_complete_df = pd.merge(student_data_df, school_data_df, on=["school_name", "school_name"])
school_data_complete_df
```

```python
student_count = school_data_complete_df["Student ID"].count()
student_count
```

```python
school_count_list = school_data_complete_df["school_name"].unique()
school_count_list
```

```python
# Calculate the total number of schools
school_list = school_data_complete_df["school_name"].unique()
school_count = len(school_list)
school_count
```

```python
# Calculate the total budget of all the scohools
Total_budget = school_data_df["budget"].sum()
Total_budget
```

```python
# Calculate the average math score for all students
Average_math = school_data_complete_df["math_score"].mean()
Average_math
```

```python
# Calculate the average reading score for all students
Average_reading = school_data_complete_df["reading_score"].mean()
Average_reading
```

```python
passing_math = school_data_complete_df["math_score"] >=70
passing_math
```

```python
#Create a new data frame of only students who passed math
passing_math = school_data_complete_df[school_data_complete_df["math_score"] >=70]
passing_math
```

```python
#Create a new data frame of only students who passed math
passing_reading = school_data_complete_df[school_data_complete_df["reading_score"] >=70]
passing_reading
```

```python
count_math_pass = passing_math["student_name"].count()
count_reading_pass = passing_reading["student_name"].count()
print(count_math_pass)
print(count_reading_pass)
```

```python
len(passing_math)
```

```python
pass_math_percentage = count_math_pass / float(student_count) * 100
pass_reading_percentage = count_reading_pass / float(student_count) * 100
print(pass_math_percentage)
print(pass_reading_percentage)
```

```python
#Filter the data frame to get both reading and math pass
passing_math_reading = school_data_complete_df[(school_data_complete_df["math_score"] >=70) & (school_data_complete_df["reading_score"] >=70)]
passing_math_reading
```

```python
overall_passing_number = passing_math_reading["student_name"].count()
overall_passing_number
```

```python
passing_math_reading_percentage = overall_passing_number / float(student_count) * 100
print(passing_math_reading_percentage)
```

```python
# Adding a list of values with keys to create a new DataFrame.
district_summary_df = pd.DataFrame(
[{"Total Schools" : school_count, "Total Students": student_count, 
  "Total Budget":Total_budget, "Average Math Score":Average_math,
  "Average Reading":Average_reading, "% Passing Math": pass_math_percentage,
  "% Passing Reading": pass_reading_percentage, "% Overall Passing":passing_math_reading_percentage}])

district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,.2f}".format)
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("{:,}".format)
district_summary_df["Average Math Score"] = district_summary_df["Average Math Score"].map("{:.2f}".format)
district_summary_df
```
